```python
4.#Write a python program to find the Radius of curvature for the curve r=4(1+cost) at t=/2
r,t=symbols('r,t')
r=4*1+cos(t)
r1=diff(r,t)
r2=diff(r1,t)
rho=(r**2+r1**2)**(3/2)/(r**2+r1**2-r*r2)
display(simplify(rho))
rho1=rho.subs(t,pi/2)
print('The radis of the curvature %0.4f units of radius %0.4f'%(pi,rho1))
```

```python
#6.Write a python program, if u = x + 3y^2 z3, v = 4x^2yz, w = 2z^2 xy then prove that at (1, 1, 0), J = 20.
x,y,z=symbols('x,y,z')
u=x+3*y**2-z**3
v=4*x**2*y*z
w=2*z**2-x*y
dux=diff(u,x)
duy=diff(u,y)
duz=diff(u,z)
dvx=diff(v,x)
dvy=diff(v,y)
dvz=diff(v,z)
dwx=diff(w,x)
dwy=diff(w,y)
dwz=diff(w,z)

J1=Matrix([[dux,duy,duz],[dvx,dvy,dvz],[dwx,dwy,dwz]])
display(J1)
J2=J1.subs([[x,1],[y,-1],[z,0]])
display(J2)
J3=Determinant(J2).doit()
display(J3)
```
